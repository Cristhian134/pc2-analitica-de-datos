```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el CSV original con las caractersticas
df = pd.read_csv('valores_caracteristicas.csv')

# Definir las caractersticas a analizar
caracteristicas = df.columns[1:-1]  # Excluyendo 'Id' y 'clase'

# Configuracin para crear grficos
for index, row in df.iterrows():
  plt.figure(figsize=(15, 10))
  
  for i, column in enumerate(caracteristicas, start=1):
    plt.subplot(3, 3, i)  # Ajustar el nmero de filas y columnas segn la cantidad de caractersticas
    sns.boxplot(data=df, x='clase', y=column)
    plt.axhline(y=row[column], color='red', linestyle='--', label='Valor Actual')
    plt.title(f'Distribucin de {column} por Clase')
    plt.xticks(rotation=45)
    plt.legend()

  plt.suptitle(f'Anlisis para el segmento: {row["Id"]}')
  plt.tight_layout(rect=[0, 0, 1, 0.96])  # Espaciado para el ttulo superior
  plt.show()
```

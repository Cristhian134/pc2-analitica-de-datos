```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el CSV original con las características
df = pd.read_csv('valores_caracteristicas.csv')

# Definir las características a analizar
caracteristicas = df.columns[1:-1]  # Excluyendo 'Id' y 'clase'

# Configuración para crear gráficos
for index, row in df.iterrows():
  plt.figure(figsize=(15, 10))
  
  for i, column in enumerate(caracteristicas, start=1):
    plt.subplot(3, 3, i)  # Ajustar el número de filas y columnas según la cantidad de características
    sns.boxplot(data=df, x='clase', y=column)
    plt.axhline(y=row[column], color='red', linestyle='--', label='Valor Actual')
    plt.title(f'Distribución de {column} por Clase')
    plt.xticks(rotation=45)
    plt.legend()

  plt.suptitle(f'Análisis para el segmento: {row["Id"]}')
  plt.tight_layout(rect=[0, 0, 1, 0.96])  # Espaciado para el título superior
  plt.show()
```
